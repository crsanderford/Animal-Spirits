### cleaning and vectorizing

```python
def clean_text(text): # Clean our tweets using regex. Remove weird characters and links.
    cleanedup = text.lower()
    return re.sub("(@[A-Za-z0-9]+)|(#[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", cleanedup)
```

```python
# load in punctuation for removal
punctuations = string.punctuation

# load in stopwords for removal
nlp = spacy.load('en_core_web_sm')
stop_words = spacy.lang.en.stop_words.STOP_WORDS

# instantiate spaCy's english parser
parser = English()
```

```python
# create a function to tokenize and lemmatize tweets, and remove stopwords and punctuation
def spacy_tokenizer(sentence):
    # Creating our token object, which is used to create documents with linguistic annotations.
    mytokens = parser(sentence)

    # Lemmatizing each token and converting each token into lowercase
    mytokens = [ word.lemma_.lower().strip() if word.lemma_ != "-PRON-" else word.lower_ for word in mytokens ]

    # Removing stop words
    mytokens = [ word for word in mytokens if word not in stop_words and word not in punctuations ]

    # return preprocessed list of tokens
    return mytokens
```

```python
# clean the tweets, divide into feature and target, divide into train and test

tweets['tweet'] = tweets['tweet'].apply(clean_text)

train, test = train_test_split(tweets, test_size=0.2, stratify=tweets['sent_score'],
                              random_state = 42)

X_train = train['tweet']
X_test = test['tweet']

y_train = train['sent_score']
y_test = test['sent_score']
```

```python
# clean and divide new tweets
newtweets_auto['tweets'] = newtweets_auto['tweets'].apply(clean_text)
newtweets_hand['tweets'] = newtweets_hand['tweets'].apply(clean_text)

X_newtweets_auto = newtweets_auto['tweets']
y_newtweets_auto = newtweets_auto['sentiment']

X_newtweets_hand = newtweets_hand['tweets']
y_newtweets_hand = newtweets_hand['sentiment']
```

```python
# initialize vectorizer for vectorizing tweets
bow_vector = CountVectorizer(tokenizer = spacy_tokenizer, ngram_range=(1,1))
```

```python
# make a pipeline for processing tweets
process_pipe = make_pipeline(
            CountVectorizer(tokenizer = spacy_tokenizer, ngram_range=(1,1)),
            SelectKBest(k=2000),
)
```

```python
# process our data
X_train_processed = process_pipe.fit_transform(X_train,y_train)
X_test_processed = process_pipe.transform(X_test)
X_newtweets_auto_processed = process_pipe.transform(X_newtweets_auto)
X_newtweets_hand_processed = process_pipe.transform(X_newtweets_hand)
```

```python
# Get the list of features that were kept
all_names = pd.Series(process_pipe.steps[0][1].get_feature_names())
selected_mask = pd.Series(process_pipe.steps[1][1].get_support())
selected_names = all_names[selected_mask]
```

```python
# re-DataFrame the data and add back in columns
X_train_kbest_df = pd.DataFrame(X_train_processed.toarray(), columns=selected_names)
X_test_kbest_df = pd.DataFrame(X_test_processed.toarray(), columns=selected_names)
X_newtweets_auto_kbest_df = pd.DataFrame(X_newtweets_auto_processed.toarray(), columns=selected_names)
X_newtweets_hand_kbest_df = pd.DataFrame(X_newtweets_hand_processed.toarray(), columns=selected_names)
```

### model training, testing, optimization

```python
%%time

ESVM = LinearSVC(C = 1.2, random_state = 42)

sum(cross_val_score(ESVM,X_train_kbest_df,y_train,cv=5))/5
```

```python
# shapley values for fake tweets

# making background data for shapley values
shapdata = shap.kmeans(X_test_kbest_df, 100)
```

```python
def process_fake_tweet(tweetstring,featurenames,processor):
    # take a string and convert it to usable format
    processed = processor.transform([tweetstring]).toarray()
    faketweet = pd.DataFrame(processed, columns=featurenames)
    return faketweet
```

```python
# instantiate the class for our explainer
explainer = shap.KernelExplainer(model=ESVM.predict, data=shapdata)
```

```python
# Write, process, and plot a fake tweet.
negative_tweet = "bitcoin is a ponzi scheme and SEC intervention is on its way"
row_process = process_fake_tweet(negative_tweet,selected_names,process_pipe)
shap_values = explainer.shap_values(row_process)

shap.initjs()
shap.force_plot(
    base_value=explainer.expected_value,
    shap_values=shap_values,
    features=row_process,
    show=False,
    matplotlib=True
)

plt.savefig('negativeshap.png')
```

```python
neutral_tweet = "bitcoin is a distributed ledger with no net moral valence"
row_process = process_fake_tweet(neutral_tweet,selected_names,process_pipe)
shap_values = explainer.shap_values(row_process)

shap.initjs()
shap.force_plot(
    base_value=explainer.expected_value,
    shap_values=shap_values,
    features=row_process
)
```
